```python
#isnull using heat map
plt.rcParams['figure.facecolor']='white'
plt.figure(figsize=(12,6))
sns.heatmap(df.isnull())
plt.title('isnull using heat map', fontsize=15)
plt.xticks(fontsize=12,color='black')
plt.yticks(fontsize=12,color='black')
plt.savefig('isnull heatmap.png',transparent=True)
plt.show()
```

# Correlation

```python
df.corr()
```

```python
plt.figure(figsize=(15,7))
sns.heatmap(df.corr(),linewidth=0.2, linecolor='Black')
plt.title('Correaltion Heatmap',fontsize=15)
plt.savefig('customer_retension_heatmap.png',transparent=True)
plt.show()
```

```python
#plotting histogram to see vizual insights
df.hist(figsize=(25,25),grid = False)
plt.savefig('customer_retension_histogram.png',transparent=True)
plt.show()
```

```python
#lets plot count plot for sex to know how may rows belongs to which sex type.
sns.countplot(df['1Gender of respondent'], palette='PuBu')
plt.title('Gender count')
plt.savefig('customer_retension_gender_count.png',transparent=True)
plt.show()
```

From above we can see that there are more female respondents(175) compared to male respondent(90)

```python
sns.countplot(df['2 How old are you? '], palette='summer')
plt.title('Age of customer')
plt.savefig('customer_retension_Age_of_customer.png',transparent=True)
plt.show()
```

most of the customers are of age betweeen 15 to 45 years old and only few are of age 10s and 50s

```python
plt.figure(figsize=(15,10))
sns.countplot(df['3 Which city do you shop online from?'], hue=df['Which of the Indian online retailer would you recommend to a friend?'],palette='Blues')
plt.title('place of shopping', fontsize=15)
plt.legend(loc='upper right', fontsize=12)
plt.xlabel('city', fontsize=12)
plt.savefig('customer_retension_place_of_shopping.png',transparent=True)
plt.show()
```

* Delhi tops the list of place of onine shopping compared to others and least number of online shoppers are from Bulandshahr and Muradabad.
* amazon and flipkart are most recommended websites in all the cities especially in Delhi, Greater Noida 

```python
plt.figure(figsize=(10,6))
sns.countplot(df['5 Since How Long You are Shopping Online ?'], palette='icefire_r')
plt.title('Since How Long You are Shopping Online', fontsize=15)
plt.xlabel('customer longitivity', fontsize=12)
plt.savefig('customer_retension_customer_longitivity.png')
plt.show()
```

* we have around 100 customers who are shopping online since 5 years and more and they the highest in numbers compared to others and the customers who are shopping since 3 years comes in second in the list.
* From this analysis we can say that online shopping has been better compared to offline shopping.

```python
plt.figure(figsize=(15,10))
sns.countplot(df['47 Getting value for money spent'], hue=df['Which of the Indian online retailer would you recommend to a friend?'],palette='YlGn')
plt.title('place of shopping', fontsize=15)
plt.legend(loc='upper right', fontsize=12)
plt.xlabel('worthy_shopping', fontsize=12)
plt.savefig('customer_retension_worthy_shopping_vs_online retailer.png',transparent=True)
plt.show()
```

* From the above outcome we can say that amazon and flipkart are the most used online retailers by the online shoppers in India and consumers rated 4 stars out of 5 for value for mopney spent.
* amazon is the most recommended online shopping website to the friends in India.
