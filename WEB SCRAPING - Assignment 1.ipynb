Q1--> python program to display all the header tags from wikipedia.org.

```python
def head_url(url):
    data={}
    
    page = requests.get(url)
    soup = BeautifulSoup(page.content)
    
    header_tags = ['h1','h2','h3']
    
    header_tag_name=[]
    header_name=[]
    for i in soup.find_all(header_tags):
        header_tag_name.append(i.name)
        header_name.append(i.text.strip())
        
    data['header_tag']=header_tag_name   
    data['header_name']=header_name
    
    return data

#making dataframe        
frame1=head_url('https://en.wikipedia.org/wiki/Main_Page')
header_tags_df=pd.DataFrame(frame1)
#DataFrame
header_tags_df
```

Q2--> python program to display IMDB’s Top rated 100 movies’ data (i.e. name, rating, year of release) and its DataFrame.

```python
def IMDB_movies_100(url):
    data = {}
    page = requests.get(url)
    soup = BeautifulSoup(page.content)

    #movie name
    name= []
    for i in soup.find_all('td',class_="titleColumn"):
        name.append(i.a.text)
    data["Name"] = name[0:100]
    
    #movie release year
    year = []
    for y in soup.find_all('span',class_="secondaryInfo"):
        year.append(y.text)
    data['year_of_release'] = year[0:100]
    
    #movie rating
    rating= []
    for r in soup.find_all('td',class_="ratingColumn imdbRating"):
        rating.append(r.strong.text)
    data[' Rating'] = rating[0:100]

    return data

frame2 = IMDB_movies_100('https://www.imdb.com/chart/top/')
# Making DataFrame
IMDB_top_100_movies = pd.DataFrame(frame2)
Ranking=np.arange(1,101)
IMDB_top_100_movies['Ranking']=Ranking
IMDB_top_100_movies.set_index('Ranking',inplace =True)

# DataFrame
IMDB_top_100_movies
```

Q4--> python program to scrape product name, price and discounts from https://meesho.com/bags-ladies/pl/p7vbp and its DataFrame

```python
def meesho(url):
    data ={}
    meeshopage = requests.get(url)
    soup = BeautifulSoup(meeshopage.content)
    
    #product name
    name =[]
    for i in soup.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
        name.append(i.text)
    data['Product_name']=name
    
    #price
    price =[]
    for i in soup.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
        price.append(i.text.replace('₹',''))
    data['Price(₹)']=price
    
    #discount
    offer =[]
    for i in soup.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH"):
        offer.append(i.text)
    data['Discount']=offer
    
    return data

#making DataFrame
frame4 = meesho('https://meesho.com/bags-ladies/pl/p7vbp')
sno = np.arange(1,17)
Meesho_products = pd.DataFrame(frame4)
Meesho_products['SNO']=sno
Meesho_products.set_index('SNO',inplace = True)

# DataFrame
Meesho_products
```

Q5---> python program to scrape cricket rankings from icc-cricket.com

  A)Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

```python
def odi_team_ranking(url):
    data={}
    
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    
    #standins
    standings=np.arange(1,11)
    data['Ranking']=standings
    
    #team name
    team = []
    for i in soup.find_all('span',class_="u-hide-phablet"):
        team.append(i.text)
    data['Team']=team[0:10]
    
    #matches
    matches1=[soup.find('td',class_="rankings-block__banner--matches").text[0:2]]
    matches = []
    for i in soup.find_all('td',class_="table-body__cell u-center-text"):
        matches.append(i.text)
    matches1.extend(matches[0:18:2])
    data['Matches']=matches1
    
    #points
    points1=[soup.find('td',class_="rankings-block__banner--points").text]
    points=[]
    for i in soup.find_all('td',class_="table-body__cell u-center-text"):
        points.append(i.text)
    points1.extend(points[1:19:2])
    data['Points']=points1
    
    #rating
    rating1=[soup.find('td',class_="rankings-block__banner--rating u-text-right").text.strip()]
    rating = []
    for i in soup.find_all('td',class_="table-body__cell u-text-right rating"):
        rating.append(i.text)
    rating1.extend(rating[0:9])
    data['Rating']=rating1
    
    return data
#Making DataFrame
frame5a =odi_team_ranking('https://www.icc-cricket.com/rankings/mens/team-rankings/odi')
Top_10_odi_team=pd.DataFrame(frame5a)
Top_10_odi_team.set_index('Ranking',inplace=True)

# DataFrame
Top_10_odi_team
```

Q7--> python program to scrape details of all the posts from coreyms.com. The heading, date, content
and the code for the video from the link for the youtube video from the post

```python
def coreyms(url):
    data = []

    page=requests.get(url)
    soup=BeautifulSoup(page.content)

    for a in soup.find_all('article'):     
        data.append({'Heading':a.h2.text,'Date':a.find('time',class_='entry-time').text,
                     'Content':a.find('div',class_='entry-content').text.strip(),
                     'Videolink':vl['src'] if (vl := a.find('iframe',class_='youtube-player')) else '-Not available-'})
        
    return data

# Making dataframe
frame7=coreyms('https://coreyms.com/')
sno=np.arange(1,11)
coreyms_DataFrame=pd.DataFrame(frame7)
coreyms_DataFrame['SNO']=sno
coreyms_DataFrame.set_index('SNO',inplace = True)

# DataFrame
coreyms_DataFrame
```

Q9--> python program to scrape mentioned details from dineout.co.in :
      i) Restaurant name
      ii) Cuisine
      iii) Location
      iv) Ratings
      v) Image URL 

```python
def dineout(url):
    data={}
    
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    
    #SNO
    sno=np.arange(1,14)
    data['SNO']=sno
    
    #restaurant name
    restaurant_name =[]
    for i in soup.find_all('div',class_="restnt-info cursor"):
        restaurant_name.append(i.a.text.strip())
    data['Restaurant_name']=restaurant_name
    
    #cuisine
    cuisine = []
    for i in soup.find_all('span',class_="double-line-ellipsis"):
        cuisine.append(i.text.replace('₹','').strip())
    data['Cuisine']=cuisine
    
    #location
    locations = []
    for i in soup.find_all('div',class_="restnt-loc ellipsis"):
        locations.append(i.text.strip())    
    data['Location']=locations
    
    #ratings
    ratings = []
    for i in soup.find_all('div',class_="restnt-rating rating-4"):
        ratings.append(i.text.strip())
    data['Ratings']=ratings
    
    #image_url
    images = []
    for i in soup.find_all('img',class_="no-img"):
        images.append(i['data-src']) # to extract image url you have to give data-src attribute as well    
    data['Image_URL']=images
    
    return data
#making DataFrame
frame9=dineout('https://www.dineout.co.in/delhi-restaurants/buffet-special')
dineout_df=pd.DataFrame(frame9)
dineout_df.set_index('SNO',inplace=True)

# DataFrame
dineout_df
```

Q10--> python program to scrape first 10 product details which include product name , price , Image URL from https://www.bewakoof.com/women-tshirts?ga_q=tshirts .

```python
def bewakoof(url):
    data={}
    
    page=requests.get(url)
    soup=BeautifulSoup(page.content)
    
    #sno
    sno=np.arange(1,11)
    data['SNO']=sno
    
    #product name
    name = []
    for i in soup.find_all('div',class_="productCardBox"):
        name.append(i.h3.text)
    data['Name']=name
    
    #price
    price = []
    for i in soup.find_all('div',class_="productPriceBox clearfix"):
        price.append(i.b.text)
    data['Price(₹)']=price
    
    #image_url
    image_url = []
    for i in soup.find_all('img',class_="productImgTag"):
        image_url.append(i['src'])
    data['Image_URL']=image_url
    
    return data

#making datafarame
frame10=bewakoof('https://www.bewakoof.com/women-printed-t-shirts')
bewakoof_df=pd.DataFrame(frame10)
bewakoof_df.set_index('SNO',inplace=True)

# DataFrame
bewakoof_df
```
